### Bonus

#### We now want to compute the mean salary after removing the maxima for each Department. Create the appropriate pivot table.

Hint: Write a custom function.

```python
# your answer here
def mean_noMax(col):
    col = col[col != col.max()].mean()
    return round(col,2)
    
employee.pivot_table(index=['Department'], aggfunc = {'Salary':mean_noMax}).reset_index()
```

```python
def mean_salary_wo_max(df):
    
    # get the index of max ID
    max_salary_indices = df.groupby('Department')['Salary'].idxmax()
    display(max_salary_indices)
    
    # drop rows that are equal to the max ID
    df_wo_max = df.drop(index=max_salary_indices)
    display(df_wo_max)
    
    # pivot table to calculate mean per department
    pt_mean_salary = df_wo_max.pivot_table(values='Salary', index='Department', aggfunc='mean')

    return pt_mean_salary

mean_salary_wo_max(employee)
```
